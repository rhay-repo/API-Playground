```python
# https://newsapi.org/account
# News API key: 75be335be615480488d812380487d1db

import requests
import json

api_key = '75be335be615480488d812380487d1db'
url = 'https://newsapi.org/v2/top-headlines?sources=techcrunch&apiKey=' + api_key

response = requests.get(url)

if response.status_code == 200:
    # The request was successful; you can now do something with the response, e.g., print it
    # print(response.json())
    print(json.dumps(response.json(), indent=2, sort_keys=True))
else:
    # The request failed
    print('Failed to retrieve data:', response.status_code)
```
